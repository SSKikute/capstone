### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import matplotlib.pyplot as plt

# Ensure that WorkWeekHrs, CodeRevHrs, and Age are loaded into the DataFrame
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Create a bubble plot of WorkWeekHrs and CodeRevHrs, using Age as bubble size
plt.figure(figsize=(10,6))

# Normalize the bubble size to make it more visually appealing
bubble_size = df['Age'].fillna(0) * 10  # Adjust this multiplier as needed

plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=bubble_size, alpha=0.5, c=df['Age'], cmap='viridis')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs (Bubble size = Age)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.colorbar(label='Age')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# Query to find the top 5 databases respondents wish to learn next year
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""
top_databases = pd.read_sql_query(QUERY, conn)

# Create a pie chart
plt.figure(figsize=(10,6))
plt.pie(top_databases['Count'], labels=top_databases['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# Query to get ConvertedComp and Age for ages 45 to 60
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df_age_comp = pd.read_sql_query(QUERY, conn)

# Calculate the median ConvertedComp for each age
median_comp_by_age = df_age_comp.groupby('Age')['ConvertedComp'].median().reset_index()

# Plot the line chart
plt.figure(figsize=(10,6))
plt.plot(median_comp_by_age['Age'], median_comp_by_age['ConvertedComp'], marker='o')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation (USD)')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# Query to count the number of respondents for each MainBranch
QUERY = """
SELECT MainBranch, COUNT(*) as Count
FROM master
GROUP BY MainBranch
"""
main_branch_df = pd.read_sql_query(QUERY, conn)

# Create a vertical bar chart with MainBranch on the X-axis and Count on the Y-axis
plt.figure(figsize=(10,6))
plt.bar(main_branch_df['MainBranch'], main_branch_df['Count'], color='skyblue')
plt.title('Number of Respondents by Main Branch')
plt.xlabel('MainBranch')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels for better readability
plt.show()

```
